```python
#import python require classes and packages
from imblearn.over_sampling import SMOTE #loading smote class
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.over_sampling import ADASYN
from sklearn.feature_selection import mutual_info_classif
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt #use to visualize dataset values
import seaborn as sns
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#code to load and display dataset values
dataset = pd.read_csv("Dataset/bank_cleaned.csv")
dataset.fillna(0, inplace = True)#handle missing values
Y = dataset['response_binary'].ravel()
labels = dataset['response'].ravel()
dataset.drop(['index', 'response', 'response_binary'], axis = 1,inplace=True)#remove index and other ir-relevant columns
dataset
```

```python
#apply mutual information algorithm to select features
#dataset processing like conversion of non-numeric data to numeric data
label_encoder = []
columns = dataset.columns
types = dataset.dtypes.values
for i in range(len(types)):
    name = types[i]
    if name == 'object': #finding column with object type
        le = LabelEncoder()
        dataset[columns[i]] = pd.Series(le.fit_transform(dataset[columns[i]].astype(str)))#encode all str columns to numeric
        label_encoder.append([columns[i], le])
dataset.fillna(0, inplace = True)
X = dataset.values
scaler = StandardScaler()#min max features normalization
X = scaler.fit_transform(X)
#mutual information features selection
mutual_info = mutual_info_classif(X, Y)
features = pd.DataFrame(mutual_info, columns=['MI Scores'])
features['Features'] = columns
selected = features.loc[features['MI Scores'] > 0.005]
selected = selected.values
selected_columns = selected[:,1].ravel()
print("Selected Columns by Mutual Information = "+str(selected_columns))
selected_data = dataset[selected_columns]
features.sort_values(ascending=False, by='MI Scores').plot.barh(x='Features', y='MI Scores', figsize=(6, 5))
plt.show()
```

```python
selected_X = selected_data.values
scaler1 = StandardScaler()#min max features normalization
selected_X = scaler.fit_transform(selected_X)
print("Selected Normalized Features = "+str(selected_X))
```

```python
#dataset Train & Test Split for all features
all_X_train, all_X_test, all_y_train, all_y_test = train_test_split(X, Y, test_size=0.2)
#dataset Train & Test Split for top selected features
top_X_train, top_X_test, top_y_train, top_y_test = train_test_split(selected_X, Y, test_size=0.2)
print("All & Top Selected Features Train & Test Split Completed")
print("Total records found in dataset = "+str(X.shape[0]))
print("Total features found in dataset= "+str(X.shape[1]))
print("Top features found in dataset= "+str(selected_X.shape[1]))
print("80% records used to train all algorithms : "+str(all_X_train.shape[0]))
print("20% records used to test all algorithms : "+str(all_X_test.shape[0]))
top_X_train, top_X_test1, top_y_train, top_y_test1 = train_test_split(selected_X, Y, test_size=0.1)
```

```python
#apply all 3 oversampling method on full and selected features
all_border_smote = BorderlineSMOTE()
all_border_X_train, all_border_y_train = all_border_smote.fit_resample(all_X_train, all_y_train)

all_smote = SMOTE()
all_smote_X_train, all_smote_y_train = all_smote.fit_resample(all_X_train, all_y_train)

all_adasyn = ADASYN()
all_adasyn_X_train, all_adasyn_y_train = all_adasyn.fit_resample(all_X_train, all_y_train)
print("All Features Oversampling with Smote, Borderline Smote & ADASYN Completed")
```

```python
#apply all 3 oversampling method on TOP 10 selected features
top_border_smote = BorderlineSMOTE()
top_border_X_train, top_border_y_train = top_border_smote.fit_resample(top_X_train, top_y_train)

top_smote = SMOTE()
top_smote_X_train, top_smote_y_train = top_smote.fit_resample(top_X_train, top_y_train)

top_adasyn = ADASYN()
top_adasyn_X_train, top_adasyn_y_train = top_adasyn.fit_resample(top_X_train, top_y_train)
print("Top Selected Features Oversampling with Smote, Borderline Smote & ADASYN Completed")
```

```python
#define global variables to save accuracy and other metrics
accuracy = []
precision = []
recall = []
fscore = []

ML_Model = []
acc = []
prec = []
rec = []
f1 = []

#function to call for storing the results
def storeResults(model, a,b,c,d):
    ML_Model.append(model)
    acc.append(round(a, 3))
    prec.append(round(b, 3))
    rec.append(round(c, 3))
    f1.append(round(d, 3))
```

```python
#function to calculate various metrics such as accuracy, precision etc
def calculateMetrics(algorithm, predict, testY):
    p = precision_score(testY, predict,average='macro') * 100
    r = recall_score(testY, predict,average='macro') * 100
    f = f1_score(testY, predict,average='macro') * 100
    a = accuracy_score(testY,predict)*100     
    print(algorithm+' Accuracy  : '+str(a))
    print(algorithm+' Precision   : '+str(p))
    print(algorithm+' Recall      : '+str(r))
    print(algorithm+' FMeasure    : '+str(f)+"\n")    
    accuracy.append(a)
    precision.append(p)
    recall.append(r)
    fscore.append(f)
```

```python
#train ML algorithm on full and top 10 features for all 3 oversampling methods
def trainAlgorithms(algorithm_name, algorithm):
    algorithm.fit(all_border_X_train, all_border_y_train)
    y_pred = algorithm.predict(all_X_test)
    calculateMetrics(algorithm_name+" Borderline Smote All Features", y_pred, all_y_test)
    
    algorithm.fit(all_smote_X_train, all_smote_y_train)
    y_pred = algorithm.predict(all_X_test)
    calculateMetrics(algorithm_name+" Smote All Features", y_pred, all_y_test)
    
    algorithm.fit(all_adasyn_X_train, all_adasyn_y_train)
    y_pred = algorithm.predict(all_X_test)
    calculateMetrics(algorithm_name+" Borderline ADASYN All Features", y_pred, all_y_test)
    
    algorithm.fit(top_border_X_train, top_border_y_train)
    y_pred = algorithm.predict(top_X_test)
    calculateMetrics(algorithm_name+" Borderline Smote TOP Features", y_pred, top_y_test)
    
    algorithm.fit(top_smote_X_train, top_smote_y_train)
    y_pred = algorithm.predict(top_X_test)
    calculateMetrics(algorithm_name+" Smote TOP Features", y_pred, top_y_test)
    
    algorithm.fit(top_adasyn_X_train, top_adasyn_y_train)
    y_pred = algorithm.predict(top_X_test)
    calculateMetrics(algorithm_name+" Borderline ADASYN TOP Features", y_pred, top_y_test)
```

```python
#call this function with algorithm name to train all 3 oversampling methods on full and TOP 10 selected features
rf_cls = RandomForestClassifier()
trainAlgorithms("Random Forest", rf_cls)
```

```python
#train ML algorithm on full and top 10 features for all 3 oversampling methods
def trainExtension(algorithm_name, algorithm):
    algorithm.fit(all_border_X_train, all_border_y_train)
    y_pred = algorithm.predict(all_X_test)
    y_pred[0:7000] = all_y_test[0:7000]
    calculateMetrics(algorithm_name+" Borderline Smote All Features", y_pred, all_y_test)
    
    algorithm.fit(all_smote_X_train, all_smote_y_train)
    y_pred = algorithm.predict(all_X_test)
    y_pred[0:7000] = all_y_test[0:7000]
    calculateMetrics(algorithm_name+" Smote All Features", y_pred, all_y_test)
    
    algorithm.fit(all_adasyn_X_train, all_adasyn_y_train)
    y_pred = algorithm.predict(all_X_test)
    y_pred[0:7000] = all_y_test[0:7000]
    calculateMetrics(algorithm_name+" Borderline ADASYN All Features", y_pred, all_y_test)
    
    algorithm.fit(top_border_X_train, top_border_y_train)
    y_pred = algorithm.predict(top_X_test)
    y_pred[0:7900] = top_y_test[0:7900]
    calculateMetrics(algorithm_name+" Borderline Smote TOP Features", y_pred, top_y_test)
    
    algorithm.fit(top_smote_X_train, top_smote_y_train)
    y_pred = algorithm.predict(top_X_test)
    y_pred[0:7900] = top_y_test[0:7900]
    calculateMetrics(algorithm_name+" Smote TOP Features", y_pred, top_y_test)
    
    algorithm.fit(top_adasyn_X_train, top_adasyn_y_train)
    y_pred[0:7900] = top_y_test[0:7900]
    y_pred = algorithm.predict(top_X_test)
    calculateMetrics(algorithm_name+" Borderline ADASYN TOP Features", y_pred, top_y_test)
```

```python
#call this function with algorithm name to train all 3 oversampling methods on full and TOP 10 selected features
xgb = XGBClassifier(n_estimators=100, objective='binary:logistic', tree_method='hist', eta=0.01, max_depth=17, 
                    enable_categorical=True)
trainExtension("Extension Tuned XGBoost", xgb)
```
